## 3. Quick Look at the Data

The starter dataset contains **30,000 content pages**.

The average number of **90-day impressions is approximately 5,200.37**, while the average number of **90-day sessions is about 37.07**.

The trend distribution shows that **16,262 pages are trending down**, which suggests there are many pages that may benefit from review or refresh. These observations make the dataset suitable for building a content opportunity scoring system.

```python
import pandas as pd

df = pd.read_csv("content_refresh_anonymized.csv")

print("Total Pages:", len(df))
print("Average 90-day Impressions:", round(df["impressions_90d"].mean(), 2))
print("Average 90-day Sessions:", round(df["sessions_90d"].mean(), 2))

print("\nTrend Direction Counts:")
print(df["trend_direction"].value_counts())
```
