#1. My Lane as an ML Task

## Lane
Refresh / Content Opportunity Scoring

## ML Task Type
Scoring and Ranking

This project is a scoring and ranking problem. The goal is not to classify pages into fixed categories but to assign each page a score based on search and engagement signals. Pages with higher scores should be reviewed first by the content team.

```python
import pandas as pd

df = pd.read_csv("content_refresh_anonymized.csv")

print("Dataset loaded successfully!")
print("Shape:", df.shape)
print("\nFirst 5 columns:")
print(df.columns[:5])
```

## 2. Target or Proxy

The dataset does not contain a direct label that tells us which pages should be refreshed. Therefore, I will use a proxy target based on observable signals such as impressions, sessions, trend direction, page freshness, and engagement.

The model will estimate a content opportunity score instead of predicting a fixed class.

```python
candidate_features = [
    "impressions_90d",
    "sessions_90d",
    "trend_direction"
]

print("Candidate features available:\n")

for feature in candidate_features:
    if feature in df.columns:
        print("✔", feature)
```

## 3. Success Metric

The success of this project will be measured by how useful the ranking is for the content team.

A good ranking should place high-opportunity pages near the top so they can be reviewed first.

Rather than maximizing prediction accuracy alone, the goal is to create a practical decision-support tool.

```python
print("Trend Direction Distribution\n")
print(df["trend_direction"].value_counts())

print("\nAverage Impressions:", round(df["impressions_90d"].mean(),2))
print("Average Sessions:", round(df["sessions_90d"].mean(),2))
```

## 4. Unit of Analysis

The unit of analysis is one content page.

Each row in the dataset represents one webpage and contains search performance, engagement, and trend information for that page.

```python
print("Unit of Analysis: One row = One content page\n")
display(df.head())
```

## 5. Why ML beats a fixed rule here

A simple rule such as "refresh every page with low traffic" ignores many important signals.

Machine learning can combine multiple features like impressions, sessions, trend direction, freshness, and engagement to identify patterns that are difficult to capture with a single rule.

The final output helps the SEO or content team prioritize their work more effectively while still allowing humans to make the final decision.

```python
print("Summary Statistics")
display(df.describe())
print("\nDataset contains", len(df), "content pages.")
```
